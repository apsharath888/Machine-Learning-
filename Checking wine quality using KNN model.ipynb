```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the data

wine = pd.read_csv("winenew.csv")
```

```python
#Taking the independent variables
x = wine.iloc[:,:-1]
x
```

```python
#taking the dependent variables
y = wine.iloc[:,-1]
y
```

```python
from sklearn import preprocessing
x =preprocessing.StandardScaler().fit_transform(x)
```

```python
x
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state =1)
```

```python
#selecting the model
from sklearn.neighbors import KNeighborsClassifier
knnmodel = KNeighborsClassifier(n_neighbors = 3)
knnmodel.fit(x_train,y_train)
ypredict = knnmodel.predict(x_test)
```

```python
#finding the accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,ypredict)
acc
```

```python
#checking the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,ypredict)
cm
```

```python
#Getting a data frame of the confusion matrix
cm = pd.DataFrame(data=cm,index=['GOOD','BAD'],columns=['GOOD','BAD'])
cm
```

```python
#Finding the values of k
ks =101
mean_acc = np.zeros((ks-1))

for n in range(1,ks):
    neigh = KNeighborsClassifier(n_neighbors=n).fit(x_train,y_train)
    y_pred = neigh.predict(x_test)
    mean_acc[n-1]=accuracy_score(y_test,y_pred)
```

```python
mean_acc
```

```python
print("The best accuracy was with",mean_acc.max()*100,"with k=",mean_acc.argmax()+1)
```

```python
#Vizualizing the accuracy of k
plt.plot(range(1,ks),mean_acc,'r')
plt.ylabel('Accuracy')
plt.xlabel('K')
plt.tight_layout()
plt.show()
```
